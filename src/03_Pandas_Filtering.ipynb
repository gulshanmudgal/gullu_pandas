```python
import pandas as pd
pd.set_option('display.max_columns', 200)
```

```python
person = {
    "first": "Gullu",
    "last": "Ji",
    "email": "gullu@test.com"
}

people = {
    "first": ["gullu", "shiv", "laxman"],
    "last": ["Ji", "Doe", "Doe"],
    "email": ["gullu@test.com", "shiv@test.com", "laxman@test.com"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
lastNameFilter = (df['last'] == "Doe")
```

```python
df[lastNameFilter]
```

```python
df[df['last'] == "Doe"] # another approach but not recommended as it's not readable
```

```python
df.loc[lastNameFilter]
```

```python
# the biggest benefit of using loc is that we can even manipulate the column list which is applicable for our case
df.loc[lastNameFilter, ['first', 'email']]
```

```python
complexFilter = (df["last"] == "Doe") & (df['first'] == "shiv")
```

```python
df.loc[complexFilter]
```

```python
complexFilter2 = (df["last"] == "Doe") | (df['first'] == "gullu")
```

```python
df.loc[complexFilter2]
```
